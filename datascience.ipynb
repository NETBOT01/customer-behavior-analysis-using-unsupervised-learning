```python
import matplotlib.pyplot as plt

plt.scatter(imputed_df['Transaction_Usage'], imputed_df['Credit_Usage'])
plt.xlabel('Transaction Usage')
plt.ylabel('Credit Usage')
plt.title('Relationship between Transaction Usage and Credit Usage')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming 'Transaction_Usage' and 'Credit_Usage' are columns in imputed_df
k_rng = range(1, 10)  # Range of k values to explore (1 to 9)
sse = []  # List to store Sum of Squared Errors (SSE)

for k in k_rng:
  # Create a KMeans model with the current k value
  km = KMeans(n_clusters=k)
  
  # Fit the model to the transaction and credit usage data (assuming these are in separate columns)
  km.fit(imputed_df[['Transaction_Usage', 'Credit_Usage']])
  
  # Append the SSE (inertia) to the list
  sse.append(km.inertia_)

# Plot the elbow graph
plt.xlabel('K (Number of Clusters)')
plt.ylabel('Sum of Squared Error (SSE)')
plt.plot(k_rng, sse)
plt.title('Elbow Plot for Transaction and Credit Usage')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the transaction and credit usage data (assuming these are in separate columns)
scaler.fit(imputed_df[['Transaction_Usage', 'Credit_Usage']])

# Transform the data using the fitted scaler
imputed_df[['Transaction_Usage', 'Credit_Usage']] = scaler.transform(imputed_df[['Transaction_Usage', 'Credit_Usage']])

# View the first few rows of the scaled DataFrame
imputed_df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(imputed_df['Transaction_Usage'], imputed_df['Credit_Usage'])
plt.xlabel('Transaction Usage')
plt.ylabel('Credit Usage')
plt.title('Relationship between Transaction Usage and Credit Usage')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

dbscan = DBSCAN(eps=0.03, min_samples=3) 
labels = dbscan.fit_predict(imputed_df[['Transaction_Usage', 'Credit_Usage']])
unique_labels = np.unique(labels)

print("Unique Cluster Labels:", unique_labels)
```

```python
import matplotlib.pyplot as plt


unique_labels = np.unique(labels)  # Get unique cluster labels

# Define a color map for multiple clusters (adjust as needed)
colors = ['blue', 'red', 'green', 'purple', 'orange', 'brown', 'pink', 'gray']

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size if desired

for label in unique_labels:
    if label == -1:  # Handle noise points (optional)
        continue
    mask = labels == label
    plt.scatter(imputed_df.loc[mask, 'Transaction_Usage'], imputed_df.loc[mask, 'Credit_Usage'],
                s=10, c=colors[label % len(colors)])  # Use modulo for color recycling

plt.xlabel('Transaction Usage')
plt.ylabel('Credit Usage')
plt.title('Cluster Visualization')
plt.legend(unique_labels[unique_labels != -1], title='Clusters')  # Exclude noise points from legend
plt.grid(True)  # Add grid lines for better readability (optional)
plt.tight_layout()  # Adjust spacing between elements
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(imputed_df[['Transaction_Usage', 'Credit_Usage']])
y_predicted
```

```python
imputed_df['cluster'] = y_predicted
imputed_df.head() 
```

```python
# Assuming your DataFrame is named 'imputed_df' and you have columns 'Transaction_Usage' and 'Credit_Usage'
km.cluster_centers_

df1 = imputed_df[imputed_df.cluster == 0]
df2 = imputed_df[imputed_df.cluster == 1]
df3 = imputed_df[imputed_df.cluster == 2]

plt.scatter(df1['Transaction_Usage'], df1['Credit_Usage'], color='green', label='Cluster 1')
plt.scatter(df2['Transaction_Usage'], df2['Credit_Usage'], color='red', label='Cluster 2')
plt.scatter(df3['Transaction_Usage'], df3['Credit_Usage'], color='black', label='Cluster 3')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='purple', marker='*', label='Centroid')

plt.xlabel('Transaction Usage')
plt.ylabel('Credit Usage')
plt.title('Cluster Visualization')
plt.legend()
plt.show()
```
